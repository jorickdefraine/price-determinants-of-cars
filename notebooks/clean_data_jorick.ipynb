```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv("../final_Cars_DB.csv")
```

```python
new_df = pd.read_csv("../final_Cars_DB.csv")
```

```python
df.info()
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.head(10)
```

```python
for row in range(len(df["price"])):
    if type(df["price"][row])==str:
        try:
            
            df["price"][row] = int(df["price"][row].replace(' ','').split("€")[0])
        except:
            df["price"][row] = 'NA'
```

```python
df.tail(25)
```

```python
for row in range(len(df["Kilométrage_compteur"])):
    if type(df["Kilométrage_compteur"][row])==str:
        try:
            df["Kilométrage_compteur"][row] = int(df["Kilométrage_compteur"][row].replace(' ','').split("Km")[0])
        except:
            df["Kilométrage_compteur"][row] = 'NA'
            
```

```python
df.head(25)
```

```python
for row in range(len(df["Boite_de_vitesse"])):
    try:
        if "mécanique" in df["Boite_de_vitesse"][row]:
            df["Boite_de_vitesse"][row] = 0
        elif "automatique" in df["Boite_de_vitesse"][row]:
            df["Boite_de_vitesse"][row] = 1
    except:
        pass
```

```python
for row in range(len(df["Boite_de_vitesse"])):
    try:
        if type(df["Boite_de_vitesse"][row])!=int:
            df["Boite_de_vitesse"][row] = 'NA'
    except:
        pass
```

```python
df.tail(25)
```

```python
for row in range(len(df["Energie"])):
    if "Essence" in df["Energie"][row]:
        df["Energie"][row] = 0
    elif "Diesel" in df["Energie"][row]:
        df["Energie"][row] = 1
    elif "lectrique" in df["Energie"][row]:
        df["Energie"][row] = 2
    else:
        df["Energie"][row] = 'NA'
```

```python
df.tail(25)
```

```python
for row in range(len(df["Nombre_de_porte"])):
    try:
        df["Nombre_de_porte"][row] = int(df["Nombre_de_porte"][row])
    except:
        df["Nombre_de_porte"][row] = "NA"
```

```python
df.tail(50)
```

```python
for row in range(len(df["Nombre_de_place"])):
    try:
        df["Nombre_de_place"][row] = int(df["Nombre_de_place"][row])
    except:
        df["Nombre_de_place"][row] = "NA"
```

```python
df.tail(50)
```

```python
test = datetime.datetime.strptime(df["Mise_en_circluation"][0].replace("/","-"), '%d-%m-%Y').date()
```

```python
for row in range(len(df["Mise_en_circluation"])):
    try:
        df["Mise_en_circluation"][row] = datetime.datetime.strptime(new_df["Mise_en_circluation"][row].replace("/","-"), '%d-%m-%Y').date()
    except:
        df["Mise_en_circluation"][row] = 'NA'
```

```python
for row in range(len(new_df["Puissance_fiscale"])):
    if type(df["Puissance_fiscale"][row])!=int:
        try:
            df["Puissance_fiscale"][row] = int(df["Puissance_fiscale"][row][:-2])
        except:
            df["Puissance_fiscale"][row] = "NA"
```

```python
df.head(50)
```

```python
for row in range(len(new_df["Puissance_din"])):
    if type(df["Puissance_din"][row])!=int:
        try:
            df["Puissance_din"][row] = int(df["Puissance_din"][row][:-2])
        except:
            df["Puissance_din"][row] = "NA"
```
